Importing the libraries for reading the dataset and necessary manupulation.

```python
import pandas as pd
import numpy as np
```

Loading the raw dataset file 1

```python
df = pd.read_csv('../data/SIMD_2020_Data.csv')
df
```

Checking Id 
1) Filling null values with 0
2) Filtering Ids with length more than 5

```python
df['id']=df['id'].fillna(value=0)
df['length_more_than_5'] = df['id'].apply(lambda x: len(str(x)) > 5)
df
```

Storing the indeices of the improper values observed in the dataset

```python
indices = df[df['length_more_than_5']].index

# Shift data one column to the right for rows where 'length_more_than_5' is True
for index in indices:
    df.iloc[index, 1:] = df.iloc[index, :-1]

missing_data_rows = df[df['id'].str.len() > 5]
missing_data_rows
```

Producing a Dataframe with filtered values

```python
cleaned_df=df[df['id'].str.len() <= 5]
cleaned_df
```

Storing it to the processed dataset folder

```python
csv_file_path = '../data/processed_data/cleaned_2020.csv'
# Writing the DataFrame to the Excel file
#cleaned_df.to_excel(excel_file_path, index=False)
cleaned_df.to_csv(csv_file_path,index=False)
```
