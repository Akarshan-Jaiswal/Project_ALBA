```python
import pandas as pd
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt

# Generate synthetic data
X, _ = make_blobs(n_samples=450, centers=3, random_state=42)

# Defining k
k=3

# Apply KMedoids
kmedoids = KMedoids(n_clusters=3, random_state=42)
kmedoids.fit(X)

# Setting cluster labels and cluster centers
labels = kmedoids.labels_
centers = X[kmedoids.medoid_indices_]

# Plot the data points and cluster centers
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k', s=50)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Medoids')
plt.title('KMedoids Clustering')
plt.legend()
plt.show()

# Reading data from file
dataset = pd.read_csv('C:/Users/Lahja/Desktop/combined_dataset_cleaned.txt')

# removing rows with Nan
dataset=dataset[dataset['Data_Zone'].notna()]
dataset

# Creating label list with numerics
labellist=numeric_cols = dataset.select_dtypes(include=['int', 'float']).columns.tolist()
labellist

# Cluster plot loops

for i in range(len(labellist)):
    for j in range(i + 1, len(labellist)):
        label1 = labellist[i]
        label2 = labellist[j]

        try:
            print('K-Medoids Clustering:', label1, label2)
            X = dataset[[label1, label2]].values
            kmedoids = KMedoids(n_clusters=k, random_state=42)
            kmedoids.fit(X)
            labels = kmedoids.labels_
            centers = X[kmedoids.medoid_indices_]

            # Plot the data points and cluster centers
            plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='plasma', s=40, edgecolors='k', marker='D')
            plt.scatter(centers[:, 0], centers[:, 1], color='red', marker='x', s=100, linewidths=2, label='Centroids')
            plt.title('K-Medoids Clustering')
            plt.xlabel(label1)
            plt.ylabel(label2)
            plt.legend()
            plt.show()
        except Exception as e:
            print("Error with:", label1, label2)
            print(e)
```
